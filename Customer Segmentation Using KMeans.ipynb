# EDA (Exploratory Data Analysis)

```python
def univariate_num(data,x):
    missing=data[x].isnull().sum()
    min1=round(data[x].min(), 2)
    max1=round(data[x].max(), 2)
    mean=round(data[x].mean(), 2)
    var=round(data[x].var(), 2)
    std=round(data[x].std(),2)
    range1=round(max1-min1, 2)
    q1=round(data[x].quantile(.25), 2)
    q2=round(data[x].quantile(.5),2)
    q3=round(data[x].quantile(.75), 2)
    skew=round(data[x].skew(), 2)
    kurt=round(data[x].kurt(), 2)
    myvalue={"missing":missing, "min":min1, "max":max1, "mean":mean,
             "var":var,"std":std, "range":range1,"q1":q1, "q2":q2,"q3":q3,
            "skewness":skew, "kurtosis":kurt}
    sns.histplot(data[x])
    plt.show()
    sns.boxplot(data=data, y=data[x])
    plt.show()
    return myvalue
```

# Outlier Treatment

```python
def myoutliers_percentile(x):
    x=x.clip(upper=x.quantile(.99))
    x=x.clip(lower=x.quantile(.01))
    return x
```

```python
data=data.apply(myoutliers_percentile)
```

## Visualizing Correlation

```python
plt.figure(figsize=(14,8))
cr=data.corr()
sns.heatmap(cr,annot=True,cmap="coolwarm")
plt.show()
```

```python
data1=pd.get_dummies(data,columns=["region"],drop_first=True)
```

# Model Development Using KMeans

```python
## selecting important features
selected_features=['voice','employ','internet','multline','reside','region_2',
            'income','retire','gender']
```

```python
data2=data1[selected_features]
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
```

```python
km_fitted=km.fit(data2)
```

```python
km_trans=km_fitted.transform(data2)
```

```python
km_fitted.inertia_ # WCSS
```

## Using Elbow Curve to find the no. of clusters to be made

```python
wcss=[]
clusters=[]
for i in range(1,20):
    km=KMeans(n_clusters=i)
    k=km.fit(data2)
    wcss.append(k.inertia_)
    clusters.append(i)
```

```python
wcss
```

```python
clusters
```

```python
## plotting the elbow curve
plt.figure(figsize=(12,5))
plt.plot(clusters,wcss)
plt.xlabel("clusters")
plt.ylabel("wcss")
plt.title("elbow curve")
plt.show()
```

from the elbow curve we can conclude that 3 clusters should be made

```python
km3=KMeans(n_clusters=3)
```

```python
km3_fitted=km3.fit(data2)
```

```python
data3=data2.copy()
```

```python
data3["cluster_3"]=km3_fitted.labels_
```

```python
data3
```

```python
data3["cluster_3"].value_counts()
```

```python
var1=pd.DataFrame(data3.groupby(["cluster_3"]).mean().T).reset_index()
var1.columns=["features","clust0","clust1","clust2"]
var1
```

```python
overall=pd.DataFrame(data2.describe().T["mean"]).reset_index()
overall.columns=["features","overallmean"]
overall
```

```python
final_clusters=pd.merge(overall,var1,on=["features"],how="inner")
```
